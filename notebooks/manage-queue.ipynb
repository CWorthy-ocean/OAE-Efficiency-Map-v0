```python
import os
from subprocess import check_output
from glob import glob

from datetime import datetime

import oae_smyle
import project
```

```python
JobIds = check_output("squeue -u ${USER} | grep run.smyl | awk '{print $1}'", shell=True).decode("utf-8").strip().split("\n")

for JobId in JobIds:
    print(JobId, end=", ")
print()
```

```python
info = {}
for JobId in JobIds:
    stdout = check_output(f"scontrol show JobId={JobId}", shell=True).decode("utf-8").strip().replace("\n", "")
    case = [s.split("=")[1].replace("run.", "") for s in stdout.split(" ") if "JobName" in s][0]
    info[case] = JobId
    
info
```

```python
cases = df_caseinfo.loc[~df_caseinfo.archive].index.to_list()
cases
```

```python
not_running = []
for case in cases:
    if case in info:
        print(f"{case}: {info[case]}")
    else:
        print(f"{case}: not running")
        not_running.append(case)

print("\nnot running:")
for case in not_running:
    log_files = sorted(glob(f"{project.dir_scratch}/{case}/run/cesm.log.????????.??????-??????"))
    if log_files: 
        log_file = log_files[-1]
    else:
        print(f"{case}: no log")
        continue
        
    with open(log_file, "r") as fid:
        lines = fid.readlines()
    CANCELLED = any(["CANCELLED" in l for l in lines])
    print(f"{case}: CANCELLED: {CANCELLED}")
    
    
    
print("\nresubmit:")
for case in not_running:    
    print(f"'{case}', ")
```
